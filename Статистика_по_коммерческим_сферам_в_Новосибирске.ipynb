# Меры центральной тенденции

## Среднее значение

###  Среднее значение количества кликов в адреса и/или входы.

```python
print(data['AddressCount'].sum()/data['AddressCount'].count())
```

###  Среднее значение количества звонков в организации.

```python
print(data['CallsCount'].sum()/data['CallsCount'].count())
```

###  Среднее значение количества кликов в организации.

```python
print(data['ClicksCount'].sum()/data['ClicksCount'].count())
```

###  Среднее значение количества организаций в той или иной сфере

```python
print(data['FirmsCount'].sum()/data['FirmsCount'].count())
```

###  Среднее значение доли трафика с карты

```python
print(data['GeoPart'].sum()/data['GeoPart'].count())
```

###  Среднее значение доли трафика с мобильных продуктов

```python
print(data['MobilePart'].sum()/data['MobilePart'].count())
```

###  Среднее значение уникальных пользователей

```python
print(data['UsersCount'].sum()/data['UsersCount'].count())
```

###  Среднее значение дистанции от пользователя до организации при совершении клика

```python
print(data['Distance'].sum()/data['Distance'].count())
```

## Медиана

```python
print("Медиана количества кликов в адреса и/или входы: {:.2f}".format(data['AddressCount'].median()))
print("Медиана количества звонков в организации: {:.2f}".format(data['CallsCount'].median()))
print("Медиана количества кликов в организации: {:.2f}".format(data['ClicksCount'].median()))
print("Медиана количества организаций в той или иной сфере: {:.2f}".format(data['FirmsCount'].median()))
print("Медиана доли трафика с карты: {:.2f}".format(data['GeoPart'].median()))
print("Медиана доли трафика с мобильных продуктов: {:.2f}".format(data['MobilePart'].median()))
print("Медиана количества уникальных пользователей: {:.2f}".format(data['UsersCount'].median()))
print("Медиана количества медианной дистанции от пользователя до организации при совершении клика: {:.2f}".format(data['Distance'].median()))
```

## Меры разброса данных
### Размах вариации

```python
print("Размах вариации количества кликов в адреса и/или входы: {:.2f}".format(data['AddressCount'].max() - data['AddressCount'].min()))
print("Размах вариации количества звонков в организации: {:.2f}".format(data['CallsCount'].max() - data['CallsCount'].min()))
print("Размах вариации количества кликов в организации: {:.2f}".format(data['ClicksCount'].max() - data['ClicksCount'].min()))
print("Размах вариации количества организаций в той или иной сфере: {:.2f}".format(data['FirmsCount'].max() - data['FirmsCount'].min()))
print("Размах вариации доли трафика с карты  : {:.2f}".format(data['GeoPart'].max() - data['GeoPart'].min()))
print("Размах вариации доли трафика с мобильных продуктов  : {:.2f}".format(data['MobilePart'].max() - data['MobilePart'].min()))
print("Размах вариации количества уникальных пользователей: {:.2f}".format(data['UsersCount'].max() - data['UsersCount'].min()))
print("Размах вариации медианной дистанции от пользователя до организации при совершении клика: {:.2f}".format(data['Distance'].max() - data['Distance'].min()))
```

### Стандартное отклонение

```python
 
print("Стандартное отклонение количества кликов в адреса и/или входы: {:.2f}".format(data['AddressCount'].std()))
print("Стандартное отклонение количества звонков в организации: {:.2f}".format(data['CallsCount'].std()))
print("Стандартное отклонение количества кликов в организации: {:.2f}".format(data['ClicksCount'].std()))
print("Стандартное отклонение количества организаций в той или иной сфере: {:.2f}".format(data['FirmsCount'].std()))
print("Стандартное отклонение доли трафика с карты: {:.2f}".format(data['GeoPart'].std()))
print("Стандартное отклонение доли трафика с мобильных продуктов: {:.2f}".format(data['MobilePart'].std()))
print("Стандартное отклонение количества уникальных пользователей: {:.2f}".format(data['UsersCount'].std()))
print("Стандартное отклонение количества медианной дистанции от пользователя до организации при совершении клика: {:.2f}".format(data['Distance'].std()))
```
